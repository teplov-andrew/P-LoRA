```python
import os
```

```python
prompts = \
   [{"prompt": live_set[i], "idx":i, "set":"live"} for i in range(len(live_set))] + \
   [{"prompt": object_set[i], "idx":i, "set":"object"} for i in range(len(live_set))]

N_gen = 15
steps = [100 * x for x in range(1, 11)]
steps
```

### Генерация изрбражений

```python
for step in steps:
    os.makedirs(f"checkpoint-{step}/sample", exist_ok=True)
    print(f"Используется чекпоинт: {step}")
    pipe = StableDiffusionXLPipeline.from_pretrained(model_id, torch_dtype=torch.float16)
    pipe = pipe.to("cuda")
    pipe.load_lora_weights(f"checkpoint-{step}/pytorch_lora_weights.safetensors", weight=1.0)

    for i in range(N_gen):
        generator = torch.manual_seed(i)
        image = pipe(
            base_prompt, 
            num_inference_steps=30, 
            guidance_scale=7.5, 
            generator=generator
        ).images[0]
        image.save(f"checkpoint-{step}/sample/base_{i}.png")
```

### Метрики

```python
from src.metrics import reconstruction_score, editability_score
from glob import glob
from tqdm import tqdm
```

```python
# reconstruction_score

rec_scores = {}
for step in tqdm(steps):
    gen_base_files = glob(f"checkpoint-{step}/sample/base_*.png", recursive=True)
    real_files = glob("/workspaces/data_train/*", recursive=True)
    
    score = reconstruction_score(gen_base_files, real_files)
    rec_scores[step] = score
    
rec_scores
```

```python
# reconstruction_score

rec_scores = {}
for step in tqdm(steps):
    gen_base_files = glob(f"checkpoint-{step}/sample/base_*.png", recursive=True)
    real_files = glob("../dataset/few_shot/instance_images/*", recursive=True)
    
    score = reconstruction_score(gen_base_files, real_files)
    rec_scores[step] = score
    
rec_scores
```

```python
# editability_score
edit_scores = {}

for step in tqdm(steps):
    all_scores = []
    
    for prompt_data in prompts:
        prompt = prompt_data['prompt']
        set_type = prompt_data['set']
        idx = prompt_data['idx']
        
        prompt = prompt.replace("sks", "")
        files = glob(f"checkpoint-{step}/sample/{set_type}_{idx}_*.png", recursive=True)
        score = editability_score(files, prompt)
        all_scores.append(score)
        
    edit_scores[step] = sum(all_scores) / len(all_scores)
        
edit_scores
```
